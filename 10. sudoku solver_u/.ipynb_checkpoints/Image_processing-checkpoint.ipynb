```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
orig_img=cv2.imread('sudoku8.jpg')
img=cv2.cvtColor(orig_img,cv2.COLOR_BGR2GRAY)
r,c=img.shape
print(r,c)
```

```python
equ=cv2.equalizeHist(img)
img=16*(img//16)
blur=cv2.GaussianBlur(img,(3,3),0)
th1=cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
th2=cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
edges1=cv2.Canny(blur,50,150)
edges2=cv2.Canny(equ,50,150)

fig,ax=plt.subplots(5,2,figsize=(15,30))

ax[0,0].imshow(img,cmap='gray')
ax[0,1].hist(img.flatten(),256,[0,256])
ax[1,0].imshow(equ,cmap='gray')
ax[1,1].hist(equ.flatten(),256,[0,256])
ax[2,0].imshow(blur,cmap='gray')
ax[2,1].hist(blur.flatten(),256,[0,256])
ax[3,0].imshow(th1,cmap='gray')
ax[3,1].imshow(th2,cmap='gray')
ax[4,0].imshow(edges1,cmap='gray')
ax[4,1].imshow(edges2,cmap='gray')

plt.show()
```

```python
orimg=th1|th2
andimg=th1&th2
_,avimg=cv2.threshold((th1+th2)//2,100,255,cv2.THRESH_BINARY)
fig,ax=plt.subplots(1,3,figsize=(20,10))
ax[0].imshow(orimg,cmap='gray')
ax[1].imshow(andimg,cmap='gray')
ax[2].imshow(avimg,cmap='gray')
plt.show()
```

```python
filter1=np.zeros((21,7),dtype=np.float32)
filter1[:,3]=1
filter1/=7

filter2=np.zeros((7,21),dtype=np.float32)
filter2[3,:]=1
filter2/=7

fig,ax=plt.subplots(1,3,figsize=(20,10))
_,inv_avimg=cv2.threshold(avimg,125,255,cv2.THRESH_BINARY_INV)
ax[0].imshow(inv_avimg,cmap='gray')

# fig,ax=plt.subplots(1,2,figsize=(10,10))
dst1=cv2.filter2D(avimg,-1,filter1)
dst2=cv2.filter2D(avimg,-1,filter2)
dst=dst1&dst2
ret,dst=cv2.threshold(dst,125,255,cv2.THRESH_BINARY_INV)
kernel=np.ones((3,3),dtype=np.uint8)
dst=cv2.dilate(dst,kernel,iterations=1)
ax[1].imshow(dst,cmap='gray')

# coords=[]
# temp=orig_img.copy()
# lines = cv2.HoughLines(dst,8,10*np.pi/180,100)
# print(len(lines))
# for i in lines:
#     rho,theta=i[0]
#     if not ((1.57<theta<1.58) or (theta<0.01) or (theta>3)):
#         continue
#     coords.append((rho,theta))
#     if len(coords)==24:
#         break
#     a = np.cos(theta)
#     b = np.sin(theta)
#     x0 = a*rho
#     y0 = b*rho
#     x1 = int(x0 + 1000*(-b))
#     y1 = int(y0 + 1000*(a))
#     x2 = int(x0 - 1000*(-b))
#     y2 = int(y0 - 1000*(a))

#     cv2.line(temp,(x1,y1),(x2,y2),(0,0,255),1)
# ax[2].imshow(temp,cmap='gray')
plt.show()
```

```python
coords=sorted(sorted(coords,key=lambda x: x[0]),key=lambda x: x[1])
print(coords,len(coords))
```

```python
y_pts=[int(i[0]) for i in coords if round(i[1])==0]
if len(y_pts)<=9:
    if y_pts[0]>c-y_pts[-1]:
        y_pts.insert(0,0)
    else:
        y_pts.append(c)
else:
    if 0 in y_pts:
        y_pts.remove(0)
    if c in y_pts:
        y_pts.remove(c)
print(y_pts, len(y_pts))
x_pts=[int(i[0]) for i in coords if round(i[1])==2]
if len(x_pts)<=9:
    if x_pts[0]>r-x_pts[-1]:
        x_pts.insert(0,0)
    else:
        x_pts.append(r)
else:
    if 0 in x_pts:
        x_pts.remove(0)
    if r in x_pts:
        x_pts.remove(r)
print(x_pts, len(x_pts))
x_width,y_width=0,0
for i,j in zip(x_pts[:-1],x_pts[1:]):
    x_width+=j-i
x_width//=(len(x_pts)-1)
for i,j in zip(y_pts[:-1],y_pts[1:]):
    y_width+=j-i
y_width//=(len(y_pts)-1)
print('Average width and height are: {} and {}'.format(x_width,y_width))
```

```python
import pickle
model=pickle.load(open('trained_ocr.pic','rb'))
print(model)
```

```python
sdk=np.zeros((9,9),dtype=np.uint8)
fig,ax=plt.subplots(9,9,figsize=(20,20))
x,y=len(y_pts),len(x_pts)
for i in zip(y_pts[:-1],y_pts[1:],range(x)):
    for j in zip(x_pts[:-1],x_pts[1:],range(y)):
        if i[0]-x_width//8<0 or i[1]+x_width//8>r or j[0]-y_width//8<0 or j[1]+y_width//8>c:
            temp=inv_avimg[i[0]:i[1],j[0]:j[1]]
        else:
            temp=inv_avimg[i[0]-x_width//8:i[1]+x_width//8,j[0]-y_width//8:j[1]+y_width//8]
        for iteration in range(2):
            x,y=temp.shape
            y1=np.where(temp[:,:y//2].sum(axis=0)//255>0.9*x)[0]
            if len(y1)==0:
                y1=0
            else:
                y1=y1.max()
            y2=np.where(temp[:,-y//2:].sum(axis=0)//255>0.9*x)[0]
            if len(y2)==0:
                y2=y
            else:
                y2=y//2+y2.min()
            x1=np.where(temp[:x//2,:].sum(axis=1)//255>0.9*y)[0]
            if len(x1)==0:
                x1=0
            else:
                x1=x1.max()
            x2=np.where(temp[-x//2:,:].sum(axis=1)//255>0.9*y)[0]
            if len(x2)==0:
                x2=x
            else:
                x2=x//2+x2.min()
            if y2-y1>20 and x2-x1>20:
                temp=temp[x1+1:x2,y1+1:y2]
        ax[i[2],j[2]].imshow(temp,cmap='gray')
        temp=cv2.resize(temp,(20,20))
        temp=temp.reshape(1,400)
        sdk[i[2],j[2]]=model.predict(temp)[0]
plt.show()
```
