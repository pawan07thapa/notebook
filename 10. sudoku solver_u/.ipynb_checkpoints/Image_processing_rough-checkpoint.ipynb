```python
import cv2
import matplotlib.pyplot as plt
```

```python
test_samples=['sudoku7.png','sudoku8.jpg']

def display_bit_plane(img):
    fig,ax=plt.subplots(2,4,figsize=(15,7))
    for i in range(2):
        for j in range(4):
            temp_img=img.copy()
            temp_img=(temp_img&2**(4*i+j))//(2**(4*i+j))
            ax[i,j].imshow(255*temp_img,cmap='gray')
    plt.show()

def load_image(img_name):
    img=cv2.imread(img_name,0)
    print(img.shape)
    return img
```

```python
img_arr=[]
for i in test_samples:
    img_arr.append(load_image(i))
for i in img_arr:
    display_bit_plane(i)
```

```python
def display_top4(img):
    temp=img.copy()
    fig,ax=plt.subplots(1,3,figsize=(15,7))
    ax[0].imshow(temp,cmap='gray')
    temp=16*(temp//16)
    ax[1].imshow(temp,cmap='gray')
    temp = cv2.medianBlur(temp,3)
    ax[2].imshow(temp,cmap='gray')
    plt.show()
    
for i in img_arr:
    display_top4(i)
```
