# 1. Visualizing the distribution of a dataset

When dealing with a set of data, often the first thing you’ll want to do is get a sense for how the variables are distributed.

## Plotting univariate distributions
The most convenient way to take a quick look at a univariate distribution in seaborn is the **`distplot()`** function. **By default, this will draw a histogram and fit a kernel density estimate (KDE)**.

To remove **kernel density estimate(KDE)**, we set `kde=False`.

**`distplot()` uses a simple rule to make a good guess for what the right number is by default, but it can be changed using `bin` parameter. For example**
```python
sns.distplot(x, bins=20, kde=False)
```

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

x = np.random.normal(size=100)
sns.distplot(x)
plt.show()

sns.distplot(x,kde=False)
plt.show()

sns.distplot(x, hist=False)
plt.show()
````

## Plotting bivariate distributions

### Jointplot
It can also be useful to visualize a bivariate distribution of two variables. **The easiest way to do this in seaborn is to just use the `jointplot()` function, which creates a multi-panel figure that shows both the bivariate (or joint) relationship between two variables along with the univariate distribution of each on separate axes.**

### Hexbin plots
The bivariate analogue of a histogram is **known as a “hexbin” plot, because it shows the counts of observations that fall within hexagonal bins. This plot works best with relatively large datasets.** It is availabel as a style in **`jointplot()` with `kind='hex'`**. It looks best with a white background.

### Kernel density estimation
It is also possible to use the kernel density estimation procedure described above to visualize a bivariate distribution. In seaborn, this kind of plot is shown with a contour plot and is available as a style in **`jointplot() with kind='kde'`**.

````python
df=pd.read_csv('iris_data.csv',names=['Petal-width','Petal-height','Sepal-width','Sepal-height','Flower'])
sns.jointplot(x="Petal-height", y="Petal-width", data=df)
plt.show()

with sns.axes_style("white"):
    sns.jointplot(x="Petal-height", y="Petal-width", kind="hex", color='k', data=df)
plt.show()

with sns.axes_style("white"):
    sns.jointplot(x="Petal-height", y="Petal-width", kind="kde", color='k', data=df)
plt.show()
````

# 2. Visualizing linear relationships

**Regression plots in seaborn are primarily intended to add a visual guide** that helps to emphasize patterns in a dataset during exploratory data analyses. That is to say that seaborn is not itself a package for statistical analysis. To obtain quantitative measures related to the fit of regression models.

Two main functions in seaborn are used to visualize a linear relationship as determined through regression. **These functions are `regplot()` and `lmplot()` are closely related. In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model and plot the resulting regression line and a 95% confidence interval for that regression**.

**`order` attribute determines the order of the polynomial which is fitted to the dataset i.e, setting `order=2` will fit quadratic ploynomial.**

````python
sns.regplot(x="Petal-width", y="Petal-height", data=df)
plt.show()
````

## Plotting a regression in other contexts

**`jointplot()` can use `regplot()` to show the linear regression fit on the joint axes by passing `kind="reg"`.**

````python
sns.jointplot(x="Sepal-width", y="Petal-width", data=df, kind="reg")
plt.show()
````

# 3. Building structured multi-plot grids

When exploring medium-dimensional data, a useful **approach is to draw multiple instances of the same plot on different subsets of your dataset**. This technique is sometimes called either **“lattice”, or “trellis” plotting**. To use these features, your data has to be in a Pandas DataFrame.

## Conditional small multiples

The FacetGrid class is useful when you want to visualize the distribution of a variable. **A FacetGrid can be drawn with up to three dimensions: row, col, and hue. The first two have obvious correspondence with the resulting array of axes; think of the hue variable as a third dimension along a depth axis, where different levels are plotted with different colors.**

For example:
```python
g = sns.FacetGrid(tips, row="smoker", col="time", margin_titles=True)
g.map(sns.regplot, "size", "total_bill", color=".3", fit_reg=False, x_jitter=.1)
```

# 4. Plotting pairwise data relationships

**PairGrid also allows you to quickly draw a grid of small subplots using the same plot type** to visualize data in each. In a PairGrid, each row and column is assigned to a different variable, so the **resulting plot shows each pairwise relationship in the dataset**. This style of plot is sometimes called a “scatterplot matrix”, as this is the most common way to show each relationship, but PairGrid is not limited to scatterplots.

The basic usage of the class is very similar to `FacetGrid`. **First you initialize the grid, then you pass plotting function to a map method and it will be called on each subplot**.  
For example:  
```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

It’s possible to plot a different function on the diagonal to show the univariate distribution of the variable in each column. For example:
```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

A very common way to use this plot colors the observations by a separate categorical variable. This is done with `hue` parameter. For example:
```python
g = sns.PairGrid(iris, hue="species")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend()
```

By default every numeric column in the dataset is used, but you can focus on particular relationships if you want, using the `var` parameter. For example:
```python
g = sns.PairGrid(iris, vars=["sepal_length", "sepal_width"], hue="species")
g.map(plt.scatter)
```

It’s also possible to use a different function in the upper and lower triangles to emphasize different aspects of the relationship. For example:
```python
g = sns.PairGrid(iris)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw=3, legend=False)
```

````python
g=sns.PairGrid(df)
g.map(plt.scatter)
plt.show()

g = sns.PairGrid(df)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
plt.show()

g = sns.PairGrid(df, hue="Flower")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend()
plt.show()

g = sns.PairGrid(df, vars=["Sepal-height", "Sepal-width"], hue="Flower")
g.map(plt.scatter)
plt.show()

g = sns.PairGrid(df, hue="Flower")
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot)
plt.show()
````

## Pairplot

`PairGrid` is flexible, but to take a quick look at a dataset, it can be easier to use **`pairplot()`**. This function uses scatterplots and histograms by default. For example:
```python
sns.pairplot(df, hue="Flower", height=2.5)
```

````python
sns.pairplot(df, hue="Flower", height=2.5)
plt.show()
````
